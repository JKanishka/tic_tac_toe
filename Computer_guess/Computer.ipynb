```python
import random

def computer_guess(x):
    low = 1
    high = x
    feedback = ''
    while feedback != 'c':
        if low != high:
            guess = random.randint(low, high)
        else:
            guess = low     # could also be high b/c low = high
        
        feedback = input(f'Is {guess} too high (H), too low (L), or correct (C) ?').lower()
        if feedback == 'h':
            high = guess - 1
        elif feedback == 'l':
            low = guess + 1
            
    print(f'yay! I Guessed {guess} it Right')


computer_guess(100)
```
